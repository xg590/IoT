```python
import pandas as pd, datetime, json
with open('sht40.log') as fr:
    idx, cols = [], []
    for i in fr.readlines():
        try:
            cols.append({j:k for j,k in json.loads(i[20:-2]).items() if j in ['Temperature (C)','Humidity (% rH)','Photoresistor']}) 
            idx.append(pd.Timestamp(i[:19]))
        except Exception as e:
            print(i, e) 
    df = pd.DataFrame(cols, index=idx)
    #axs = df[['Temperature (C)','Photoresistor']].plot.line(figsize=(16, 14), subplots=True)
```

```python
df['Temperature (C)'].plot.line(figsize=(16, 10)) 
```
